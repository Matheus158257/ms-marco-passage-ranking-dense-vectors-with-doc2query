```python
#Bibliotecas Padrão
import os
import csv
import json
import pdb
import pandas as pd
```

```python
csv_file =  open(ids_path_dev, encoding="cp1252",mode = 'r', errors='ignore')
csv_reader = csv.reader(csv_file, delimiter='\t')
df = pd.DataFrame(csv_reader,columns=['qid','pid','question','passage'])
```

```python
list_records= df.to_dict('records')
list_records
```

```python
dict_uniques = {}
# for qid_unique in qid_uniques:
for elem in list_records:
    try:
        dict_uniques[elem['qid']].append(elem['pid'])
    except:
        dict_uniques[elem['qid']]= [elem['pid']]
dict_ordered = {}
i=0
for item in dict_uniques.items():
    dict_ordered[str(i)] = item
    i+=1
print(i)
with open('inference_data_dev_top1000.json', 'w') as f:
    json.dump(dict_ordered, f)
```
